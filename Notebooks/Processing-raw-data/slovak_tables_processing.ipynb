# Slovak data

```python
import pandas as pd
from IPython.display import Markdown;
import numpy
```

## slovak_immigration.csv

```python
df = pd.read_csv("../../Data/slovakia/slovak_immigration.csv")
country_codes = pd.read_csv('../../Data/country_region_subregion_codes.csv')

merged_df = df.merge(country_codes, left_on='citizen', right_on='alpha-2', how='left')

merged_df['geo'] = 'Slovakia'
merged_df['Geopolitical entity (reporting)'] = 'SVK'

merged_df = merged_df.drop(columns=['citizen', 'name', 'alpha-2', 'country-code','iso_3166-2','intermediate-region','region-code','sub-region-code','intermediate-region-code', 'STRUCTURE', 'STRUCTURE_ID', 'STRUCTURE_NAME', 'freq', 'Time frequency', 'agedef', 'Age definition', 'Age class', 'unit', 'Unit of measure', 'sex', 'Time'])
merged_df = merged_df.rename(columns={'Country of citizenship': 'citizen_country', 'alpha-3': 'citizen_ISO3', 'region': 'citizen_region','sub-region': 'citizen_sub_region', 'geo': 'country', 'Geopolitical entity (reporting)': 'country_ISO3', 'TIME_PERIOD': 'year', 'OBS_VALUE': 'count', 'Sex': 'sex'})

columns_order = list(merged_df.columns)
columns_order.remove('citizen_ISO3')
columns_order.remove('citizen_region')
columns_order.remove('citizen_sub_region')
columns_order.insert(1, 'citizen_ISO3')
columns_order.insert(2, 'citizen_region')
columns_order.insert(3, 'citizen_sub_region')
merged_df = merged_df[columns_order]

merged_df = merged_df.iloc[:, :-5]

display(merged_df.head())

file_save_path = "../../Data/slovakia/slovak_immigration.csv"
merged_df.to_csv(file_save_path,index=False)


```
