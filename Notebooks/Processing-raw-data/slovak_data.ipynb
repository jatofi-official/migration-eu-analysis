# This notebook edits slovak district data

### Importing libraries

```python
import pandas as pd
import geopandas
```

### Importing tables

```python
immigration_table_raw = pd.read_csv("../../Data/slovakia/immigration_districts_sk.csv")
emigration_table_raw = pd.read_csv("../../Data/slovakia/emigration_districts_sk.csv")

pop_change_table_raw = pd.read_csv("../../Data/slovakia/population_change_sk.csv")


display(immigration_table_raw.head())
display(emigration_table_raw.head())
display(pop_change_table_raw.head())

```

#### Calculating table dimensions

```python
n_cols = len(pop_change_table_raw.iloc[0,:])
n_rows = len(pop_change_table_raw.iloc[:,0])

print("Rows: ",n_rows)
print("Columns: ",n_cols)
```

Calculating net migration

```python
net_migration_table = immigration_table_raw.copy()

for i in range(n_rows):
    for k in range(1,n_cols):
        net_migration_table.iloc[i,k] = immigration_table_raw.iloc[i,k]-emigration_table_raw.iloc[i,k]

display(net_migration_table.head())

net_migration_table.to_csv("../../Data/slovakia/net_migration_districts_sk.csv",index=False)
```
