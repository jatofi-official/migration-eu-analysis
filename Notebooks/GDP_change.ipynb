### Importing libraries


```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import Markdown
import geopandas
import plotly.express as px
import numpy
import seaborn as sns
```

### Setting up the table and plotting

```python
GDP_table=pd.read_csv("../Data/GDP.csv")
y_range=max(GDP_table["OBS_VALUE"].min(),GDP_table["OBS_VALUE"].max())
for state in GDP_table["Geopolitical entity (reporting)"].unique():
    country_df=GDP_table[GDP_table["Geopolitical entity (reporting)"]==state]
    y_values=range(int(-y_range),int(y_range),4)
    years = sorted(GDP_table["TIME_PERIOD"].unique())
    plt.plot(country_df["TIME_PERIOD"],country_df["OBS_VALUE"])
    plt.title(f"GDP of {state}")
    plt.ylim(-y_range,y_range)
    plt.yticks(y_values)
    plt.xticks(years, rotation=45)
    plt.grid()
    plt.show()
```
